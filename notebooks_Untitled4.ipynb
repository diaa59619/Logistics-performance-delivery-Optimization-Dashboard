```python
np.random.seed(42)
random.seed(42)

n = 25000

cities = ["Delhi", "Mumbai", "Bangalore", "Chennai", "Kolkata", "Hyderabad", "Ahmedabad", "Pune"]
shipping_modes = ["Air", "Road", "Rail"]
customer_types = ["Corporate", "Retail"]

start_date = datetime(2023, 1, 1)

data = []

for i in range(n):
    order_date = start_date + timedelta(days=random.randint(0, 364))
    dispatch_delay = random.randint(0, 2)
    dispatch_date = order_date + timedelta(days=dispatch_delay)
    
    mode = random.choice(shipping_modes)
    
    if mode == "Air":
        transit_days = random.randint(1, 3)
    elif mode == "Road":
        transit_days = random.randint(3, 7)
    else:
        transit_days = random.randint(4, 8)
    
    expected_delivery = dispatch_date + timedelta(days=transit_days)
    actual_delay = random.choice([0, 0, 0, 1, 2, 3])
    delivery_date = expected_delivery + timedelta(days=actual_delay)
    
    distance = random.randint(100, 2500)
    weight = round(random.uniform(1, 50), 2)
    
    revenue = round(distance * weight * 0.05 + random.randint(500, 2000), 2)
    cost = round(revenue * random.uniform(0.6, 0.85), 2)
    
    delivery_status = "Delayed" if actual_delay > 0 else "On-Time"
    
    data.append([
        f"SHP{i+1}",
        order_date,
        dispatch_date,
        delivery_date,
        expected_delivery,
        random.choice(cities),
        random.choice(cities),
        mode,
        distance,
        weight,
        revenue,
        cost,
        random.choice(customer_types),
        delivery_status
    ])

columns = [
    "Shipment_ID", "Order_Date", "Dispatch_Date", "Delivery_Date",
    "Expected_Delivery_Date", "Origin_City", "Destination_City",
    "Shipping_Mode", "Distance_km", "Weight_kg",
    "Revenue", "Cost", "Customer_Type", "Delivery_Status"
]

df = pd.DataFrame(data, columns=columns)

print("Total rows:", len(df))
df.head()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Dispatch_Date'] = pd.to_datetime(df['Dispatch_Date'])
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'])
df['Expected_Delivery_Date'] = pd.to_datetime(df['Expected_Delivery_Date'])
```

```python
# Profit
df['Profit'] = df['Revenue'] - df['Cost']

# Delay Days
df['Delay_Days'] = (df['Delivery_Date'] - df['Expected_Delivery_Date']).dt.days

# Actual Delivery Time
df['Delivery_Time_Days'] = (df['Delivery_Date'] - df['Dispatch_Date']).dt.days

# Month
df['Month'] = df['Order_Date'].dt.month

# On-Time Flag (Numeric)
df['On_Time_Flag'] = df['Delivery_Status'].apply(lambda x: 1 if x == "On-Time" else 0)

# Profit Margin
df['Profit_Margin'] = df['Profit'] / df['Revenue']

# Route
df['Route'] = df['Origin_City'] + " → " + df['Destination_City']

# Year-Month
df['Year_Month'] = df['Order_Date'].dt.to_period('M').astype(str)

df.head()
```
